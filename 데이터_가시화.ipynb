우선 임의의 활력징후 신호를 먼저 가시화





> **심박**









```
# 축 설정
x=xls_data.loc[:,['time']]
y=xls_data.loc[:,['heart_rate']]
```

```
# 점으로 표현
plt.title("heart rate")
plt.xlabel('second') 
plt.ylabel('count')
plt.scatter(x, y)
plt.show()
# 선으로 표현
plt.title("heart rate")
plt.xlabel('second') 
plt.ylabel('count')
plt.plot(x, y)
plt.show()
```

> **호흡수**

```
# 축 설정
x=xls_data.loc[:,['time']]
y=xls_data.loc[:,['breathing_rate']]
```


**활력징후 신호 전체 그래프**

```
plt.title("signal")
plt.xlabel('second') 
plt.ylabel('count')
plt.plot(xls_data.time, xls_data.heart_rate, color='g', ls=':', label='heart rate') 
plt.plot(xls_data.time, xls_data.breathing_rate, color='r', ls='-', label='breathing rate')
plt.plot(xls_data.time, xls_data.body_temperature, color='b', ls='-', label='body temperature')
plt.legend(shadow=True, fancybox=True, loc="upper left")
```

## 데이터 과학
</div>

Colab을 통해 인기 있는 Python 라이브러리를 최대한 활용하여 데이터를 분석하고 시각화할 수 있습니다. 아래 코드 셀에서는 <strong>Numpy</strong>를 사용하여 임의의 데이터를 생성하고 <strong>매트플롯립</strong>으로 이를 시각화합니다. 셀을 클릭하면 코드 수정을 바로 시작할 수 있습니다.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```
